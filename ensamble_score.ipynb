```
# turn masked model labelling to data labelling format
masked_prediction_final = {}
for i in range(len(test_labels)):
  if len(masked_prediction_results[i]) !=3:
    masked_prediction_final[i] = ["NA", "NA", "NA"]
    continue
  else:
    if masked_prediction_results[i][2] == 1:
      # the prediction for ith sample is correct
      if test_labels[i] == 0:
        masked_prediction_final[i] = [masked_prediction_results[i][0],masked_prediction_results[i][1],test_labels[i]]
      elif test_labels[i] == 1:
        masked_prediction_final[i] = [masked_prediction_results[i][1],masked_prediction_results[i][0],test_labels[i]]
    elif masked_prediction_results[i][2] == 0:
      # the prediction for ith sample is wrong
      if test_labels[i] == 0:
        masked_prediction_final[i] = [masked_prediction_results[i][1],masked_prediction_results[i][0],1-test_labels[i]]
      elif  test_labels[i] == 1:
        masked_prediction_final[i] = [masked_prediction_results[i][0],masked_prediction_results[i][1],1-test_labels[i]]
    else:
      # the results cannot be computed
      masked_prediction_final[i] = ["NA", "NA", "NA"]
```

```
# turn masked model labelling to data labelling format
MA_prediction_final = {}
for i in range(len(test_labels)):
  if len(MA_results[i]) !=3:
    MA_prediction_final[i] = ["NA", "NA", "NA"]
    continue

  else:
    if MA_results[i][2] == 1:
      # the prediction for ith sample is correct
      if test_labels[i] == 0:
        MA_prediction_final[i] = [MA_results[i][0],MA_results[i][1],test_labels[i]]
      elif test_labels[i] == 1:
        MA_prediction_final[i] = [MA_results[i][1],MA_results[i][0],test_labels[i]]
    elif MA_results[i][2] == 0:
      # the prediction for ith sample is wrong
      if test_labels[i] == 0:
        MA_prediction_final[i] = [MA_results[i][1],MA_results[i][0],1-test_labels[i]]
      elif  test_labels[i] == 1:
        MA_prediction_final[i] = [MA_results[i][0],MA_results[i][1],1-test_labels[i]]
    else:
      # the results cannot be computed
      MA_prediction_final[i] = ["NA", "NA", "NA"]
```

```
MA_prediction_final
```

```
score_MA = {}
for i in range(len(test_labels)):
  if MA_prediction_final[i][-1] != "NA":
    score_MA[i] = [MA_prediction_final[i][0],MA_prediction_final[i][1]]
  else:
    score_MA[i] = ["NA","NA"]
```

```
# ensemble
ensemble_score = {}
for i in range(len(test_labels)):
  print(i)
  pair_score_sent0 = 1-score_pair_classification[i][0]  # for being correct
  pair_score_sent1 = 1-score_pair_classification[i][1]
  mask_score_sent0 = score_masked_model[i][0]
  mask_score_sent1 = score_masked_model[i][1]
  MA_score_sent0 = score_MA[i][0]
  MA_score_sent1 = score_MA[i][1]
  if i == 248:
    ensemble_score[i] = [np.mean([pair_score_sent0,mask_score_sent0]),np.mean([pair_score_sent1,mask_score_sent1])]
  elif i == 1020:
    ensemble_score[i] = [pair_score_sent0,np.mean([pair_score_sent1,MA_score_sent1])]
  elif i == 1539:
    ensemble_score[i] = [np.mean([pair_score_sent0,mask_score_sent0]),np.mean([pair_score_sent1,mask_score_sent1])]
  else:
    sent0_score = np.mean([pair_score_sent0,mask_score_sent0,MA_score_sent0])
    sent1_score = np.mean([pair_score_sent1,mask_score_sent1,MA_score_sent1])
    ensemble_score[i] = [sent0_score,sent1_score]
```

```
ensemble_score
```

```
ensemble_prediction = []
for i in range(len(test_labels)):
  score_list = ensemble_score[i]
  ensemble_prediction.append(score_list.index(min(score_list)))
```

```
sum(ensemble_prediction == test_labels)/2000
```
